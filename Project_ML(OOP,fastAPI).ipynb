```python
class CoursecompletionModel:
    def __init__(self):
        with open(model.pkl,'rb') as f:
            self.model = pickle.load(f)
    def predict(self, features_dict):
        feature_list = [
            features_dict['age'],
            features_dict['hours_per_week'],
            features_dict['num_logins_last_month'],
            features_dict['videos_watched_pct'],
            features_dict['assignments_submitted'],
            features_dict['discussion_posts'],
            features_dict['is_working_professional'],
            features_dict['height_cm'],
            features_dict['weight_kg'],
            features_dict['num_siblings'],
            features_dict['has_pet'],
            features_dict['desktop'],
            features_dict['mobile'],
            features_dict['pager'],
            features_dict['smart_tv'],
            features_dict['tablet']
        ]
        features_array = np.array([feature_list])
        result = self.model.predict(features_array)[0]  # pick first element
        return int(result)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class StudentInput(BaseModel):
    age: float
    hours_per_week: float
    num_logins_last_month: int
    videos_watched_pct: float
    assignments_submitted: int
    discussion_posts: int
    is_working_professional: int
    height_cm: float
    weight_kg: float
    num_siblings: int
    has_pet: int
    desktop: int
    mobile: int
    pager: int
    smart_tv: int
    tablet: int

@app.post("/predict_course")
def predict_course(data: StudentInput):
    features_dict = data.dict()
    pred = model.predict(features_dict)
    if pred == 1:
        status = "Completed" 
    else:
        "Not Completed"
    
    return {"student_prediction": status}
```
